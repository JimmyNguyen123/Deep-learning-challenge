## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import h5py

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df=application_df.drop(['EIN','NAME'],axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace=['T9','T13','T12','T2','T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
values_count = application_df['CLASSIFICATION'].value_counts()
filtered_values=values_count[values_count>1]
print(filtered_values)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
for cls in values_count.index:
    if cls not in ['C1000','C2000','C1200','C3000','C2100']:
        classifications_to_replace.append(cls)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

#Convert and rename APPLICATION_TYPE
app_type_d=pd.get_dummies(application_df['APPLICATION_TYPE'])
app_type_d=app_type_d.rename(columns={'Other':'Application Type (Other)'})

#Convert and rename AFFILIATION
aff_d = pd.get_dummies(application_df['AFFILIATION'])
aff_d=aff_d.rename(columns={'Other':'Affiliation (Other)'})

#Convert and rename CLASSIFICATION
cls_d = pd.get_dummies(application_df['CLASSIFICATION'])
cls_d = cls_d.rename(columns={'Other':'Classification (Other)'})

#Convert and rename USE_CASE
usecase_d = pd.get_dummies(application_df['USE_CASE'])
usecase_d=usecase_d.rename(columns={'Other':'Use Case (Other)'})

#Convert ORGANIZATION
org_d = pd.get_dummies(application_df['ORGANIZATION'])

#Convert and rename INCOME_AMT
income_d = pd.get_dummies(application_df['INCOME_AMT'])
income_d = income_d.rename(columns={'0':'Income = 0'})

#Convert SPECIAL_CONSIDERARIONS to 1 and 0
for value in application_df['SPECIAL_CONSIDERATIONS']:
    if value == 'N':
        application_df['SPECIAL_CONSIDERATIONS']=application_df['SPECIAL_CONSIDERATIONS'].replace(value,0)
    else:
        application_df['SPECIAL_CONSIDERATIONS']=application_df['SPECIAL_CONSIDERATIONS'].replace(value,1)
```

```python
#Drop columns before concat dummy columns
application_df=application_df.drop(['APPLICATION_TYPE',
                                    'AFFILIATION',
                                    'CLASSIFICATION',
                                    'USE_CASE',
                                    'ORGANIZATION',
                                    'INCOME_AMT'],axis=1)
application_df.head()
```

```python
#Combine application_df with dummy columns
application_df=pd.concat([application_df,
                         app_type_d,
                         aff_d,
                         cls_d,
                         usecase_d,
                         org_d,
                         income_d
                         ],axis=1)
```

```python
#Check final application_df
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y=application_df['IS_SUCCESSFUL'].values
X=application_df.drop(['IS_SUCCESSFUL'],axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```
